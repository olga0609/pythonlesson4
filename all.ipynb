```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
# пункт 1 - найти владельца документа
def search_number():
    user_number = input('Введите номер документа: ')
    owner=""
    for item in documents:
        if item.get('number') == user_number:
            owner = item.get('name')
    if owner != "":
        print(f'Владелец документа - {owner}')
    else:
        print('Документ не найден в базе')
        

# пункт 2 - вывести номер полки
def search_shelf():
    user_number = input('Введите номер документа: ')
    shelf = ""
    for num in directories:
        for value in directories[num]:
            if value == user_number or value == number:
                shelf = num
    if shelf != "":
        print(f'Документ хранится на полке: {shelf}')
    else:
        print('Документ не найден в базе')

# пункт 3 - вывести информацию по всем документам
def documents_information():
    for item in documents:
        number = (((item['number'])))
        for key, value in directories.items():
            if number in value:
                number = item['number']
                type = item['type']
                owner = item['name']
                shelf = key
                print(f'№: {number}, тип: {type}, владелец: {owner}, полка хранения:{shelf}')
            
# пункт 4 - добавить новую полку
def add_shelf():
    new_item = input('Введите номер полки: ')
    if new_item in directories.keys():
        print(f'Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}')
    else:
        directories[new_item] = ""
        print(f'Полка добавлена. Текущий перечень полок: {list(directories.keys())}')

# пункт 5 - удалить полку если она пустая
def delete_shelf():
    shelf = input('Введите номер полки: ')
    if shelf in directories.keys() and len(directories[shelf]) == 0:
        del directories[shelf]
        print(f'Полка удалена. Текущий перечень полок: {list(directories.keys())}')
    elif shelf in directories.keys() and len(directories[shelf]) != 0:
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {list(directories.keys())}')

command = input('Введите команду: ')
while True:
    command = input('Введите команду: ')
    if command == 'p':
        search_number()
    elif command == 's':
        search_shelf()
    elif command == 'l':
        documents_information()
    elif command == 'ads':
        add_shelf()
    elif command == 'ds':
        delete_shelf()
    elif command == 'q':
        break
```
